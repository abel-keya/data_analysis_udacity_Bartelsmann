```
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis');
```

```

# import labelencoder
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
lb_make = LabelEncoder()
# apply le on categorical feature columns
data['Gender'] = lb_make.fit_transform(data['Gender'])
data['Neighbourhood'] = lb_make.fit_transform(data['Neighbourhood'])
data
```

```
data['ScheduledDay'] = pd.to_datetime(data['ScheduledDay'],format='%Y-%m-%d %H:%M:%S').dt.tz_localize(None)
data['AppointmentDay'] = pd.to_datetime(data['AppointmentDay'],format='%Y-%m-%d').dt.tz_localize(None)
data
```

```
data['ScheduledDay'] = pd.to_datetime(data['ScheduledDay'],format='%Y-%m-%d %H:%M:%S').dt.tz_localize(None)
data['AppointmentDay'] = pd.to_datetime(data['AppointmentDay'],format='%Y-%m-%d').dt.tz_localize(None)
#Extract AppointmentDay to a new column
data['appointmentweek'] = data['AppointmentDay'].dt.week
data['appointmentmonth'] = data['AppointmentDay'].dt.month
data['appointmentyear'] = data['AppointmentDay'].dt.year
data['appointmentday'] = data['AppointmentDay'].dt.day
data['scheduledtweek'] = data['ScheduledDay'].dt.week
data['scheduledmonth'] = data['ScheduledDay'].dt.month
data['scheduledyear'] = data['ScheduledDay'].dt.year
data['scheduledday'] = data['ScheduledDay'].dt.day
data['scheduledhour'] = data['ScheduledDay'].dt.hour
data['scheduledmin'] = data['ScheduledDay'].dt.minute
data
```
